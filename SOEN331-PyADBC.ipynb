## Functions

$$ factorial(n) = \begin{cases}
1 &\mbox{if } n = 0 \\ 
n \times factorial(n-1) & \mbox{if } n > 1. \end{cases} $$

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print factorial(5)
```

## Decorators!

**Decorators** are **functions** that produce cooler **functions**.

This function `wrap()` makes a function print a message every time it is called.

```python
def wrap(fun):
    def newfun(*args, **kwargs):
        print 'calling the function'
        return fun(*args, **kwargs)
    return newfun

@wrap
def fac(n):
    if n == 0:
        return 1
    else:
        return n * fac(n-1)

print fac(5)
```

## Lambdas

```python
def fac(n):
    return n == 0 and 1 or n * fac(n-1)

fac = lambda n: n == 0 and 1 or n * fac(n-1)
fac(5)
```

# Basic Usage

## `@invariant`

```python
@invariant(lambda self: self.capacity >= 0)
class List(object):
    pass
```

## `@requires` for pre-conditions

```python
class List(object):
    # ...
    @requires(lambda self: self.size() < self.capacity)
    def append(self, bla):
        pass
```

## `@ensures` for post-conditions

```python
class List(object):
    # ...    
    @ensures(lambda self: self.size() == len(self.things))
    def append(self, bla):
        self._size += 1
        self.things.append(bla)
```

## But what about detecting changes? (`$old()` in adbc)

This is a bit of a hack...

```python
    @ensures(lambda self, old: self.size() == old['size'] + 1)
    @old(lambda self: {'size': self.size()})
    def append(self, bla):
        self._size += 1
        self.things.append(bla)
```

## A LIST

```python
@invariant(lambda self: self.capacity >= 0)
class List(object):
    def __init__(self, capacity):
        self.capacity = capacity
        self.things = []
        self._size = 0

    @requires(lambda self: self.size() < self.capacity)
    @ensures(lambda self, old: self.size() == old['size'] + 1)
    @old(lambda self: {'size': self.size()})
    def append(self, bla):
        self._size += 1
        self.things.append(bla)

    @requires(lambda self: self.size() < self.capacity)
    @ensures(lambda self, old: self.size() == old['size'] + 1)
    @old(lambda self: {'size': self.size()})
    def append_buggy(self, bla):
        self.things.append(bla)

    def size(self):
        return self._size
```

```python
li = List(5)
for i in range(10):
    print "appending ", i
    li.append(i)
```

```python
li = List(5)
li.append_buggy(1)
```
